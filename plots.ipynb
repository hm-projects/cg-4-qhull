```python
import pandas as pd
```

```python
POINTS = "Points (X)"
DIMENSION = "Dimension (Y)"
TIMING = "Timing"
DISTANCE_TEST = "Distance tests for qhull"
```

```python
sphere = pd.read_csv("sphere.csv")
sphere = sphere.set_index(POINTS)
sphere.plot(subplots=True)
```

```python
spiral = pd.read_csv("spiral.csv")
spiral = spiral.set_index(POINTS)
spiral.plot(subplots=True)
```

```python
cube = pd.read_csv("cube.csv")
cube = cube.set_index(DIMENSION)

dimensions = [2,3,4,5,6,7,8]

for d in dimensions:
    filtered = cube.filter(like="{d}".format(d=d),axis=0)
    filtered = filtered.reset_index()
    filtered.drop(DIMENSION,inplace=True,axis=1)
    filtered = filtered.set_index(POINTS)
    filtered.plot(subplots=True,title="Dimension {d}".format(d=d))
```
