```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
POINTS = "Points (X)"
DIMENSION = "Dimension (Y)"
TIMING = "Timing (s)"
DISTANCE_TEST = "Distance tests for qhull"
```

```python
sphere = pd.read_csv("sphere.csv")
sphere = sphere.set_index(POINTS)
_ = sphere.plot(
    subplots=True, title="Sphere QHULL Distance Tests and Timing (s) Change by Deminsion and Number of Points")
```

```python
spiral = pd.read_csv("spiral.csv")
spiral = spiral.set_index(POINTS)
_ = spiral.plot(subplots=True,
                title="Spiral QHULL Distance Tests and Timing (s) Change by Deminsion and Number of Points")
```

```python
cube = pd.read_csv("cube.csv")

c_timing = cube.drop(DISTANCE_TEST, axis=1)
c_distance = cube.drop(TIMING, axis=1)

_ = c_distance.groupby([POINTS, DIMENSION]).first().unstack().plot(
    title="Cube QHULL Distance Tests Change by Deminsion and Number of Points")
_ = c_timing.groupby([POINTS, DIMENSION]).first().unstack().plot(
    title="Cube QHULL Timing (s) Change by Deminsion and Number of Points")
```
